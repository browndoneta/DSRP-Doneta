```python
df.head()
df.info()
df.describe()
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
df = df[df['Age'] > 120]
```

```python
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
```

```python
df.isnull().sum()
```

```python
df.to_csv("cleaned_data.csv", index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize=(6, 4))
ax = sns.countplot(x='Gallstone Status', data=df)


for p in ax.patches:
 height = p.get_height()
ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2., height),
        ', fontsize=10)


plt.title("Gallstone Disease Prevalence")
plt.xlabel("Gallstone Status")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
