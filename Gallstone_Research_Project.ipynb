```python
import statsmodels.api as sm
import pandas as pd

# Metabolic indicators only
metabolic_indicators = [
'Glucose',
'Total Cholesterol (TC)',
'Low Density Lipoprotein (LDL)',
'High Density Lipoprotein (HDL)',
'Triglyceride',
'Aspartat Aminotransferaz (AST)',
'Alanin Aminotransferaz (ALT)',
'Alkaline Phosphatase (ALP)',
'Creatinine'
]


df_model = df[metabolic_indicators + ['Gallstone Status']].dropna()

X = df_model[metabolic_indicators]
y = df_model['Gallstone Status']


X = sm.add_constant(X)


logit_model = sm.Logit(y, X).fit()


print(logit_model.summary())
```

Are metabolic indicators significantly associated with the presence of gallstone disease?

```python
# Odds Ratios
import numpy as np

odds_ratios = pd.DataFrame({
"Variable": X.columns,
"Odds Ratio": np.exp(logit_model.params),
"P-Value": logit_model.pvalues
})

print(odds_ratios.sort_values("P-Value"))
```

```python
# Body composition indicators only
body_comp_indicators = [
'Weight',
'Body Mass Index (BMI)',
'Total Body Fat Ratio (TBFR) (%)',
'Visceral Fat Rating (VFR)',
'Visceral Fat Area (VFA)',
'Lean Mass (LM) (%)',
'Muscle Mass (MM)',
'Total Fat Content (TFC)',
'Obesity (%)',
'Body Protein Content (Protein) (%)',
'Bone Mass (BM)',
'Hepatic Fat Accumulation (HFA)',
'Extracellular Fluid/Total Body Water (ECF/TBW)'
]


import statsmodels.api as sm
import numpy as np
import pandas as pd

df_body = df[body_comp_indicators + ['Gallstone Status']].dropna()


X = df_body[body_comp_indicators]
y = df_body['Gallstone Status']


X = sm.add_constant(X)


logit_model_body = sm.Logit(y, X).fit()


print(logit_model_body.summary())
```

```python
odds_ratios_body = pd.DataFrame({
"Variable": X.columns,
"Odds Ratio": np.exp(logit_model_body.params),
"P-Value": logit_model_body.pvalues
})

print(odds_ratios_body.sort_values("P-Value"))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


odds_ratios = np.exp(logit_model.params)
conf = logit_model.conf_int()
conf_odds = np.exp(conf)
p_values = logit_model.pvalues


results_df = pd.DataFrame({
'Variable': odds_ratios.index,
'Odds Ratio': odds_ratios.values,
'CI Lower': conf_odds[0].values,
'CI Upper': conf_odds[1].values,
'P-Value': p_values.values
})


results_df = results_df.sort_values(by='Odds Ratio', ascending=False)

# Plot
plt.figure(figsize=(8, 6))
sns.pointplot(data=results_df, y='Variable', x='Odds Ratio', join=False, color='blue')
plt.errorbar(x=results_df['Odds Ratio'], y=results_df['Variable'],
xerr=[results_df['Odds Ratio'] - results_df['CI Lower'],
results_df['CI Upper'] - results_df['Odds Ratio']],
fmt='none', ecolor='gray', capsize=3)

plt.axvline(x=1, color='red', linestyle='--')
plt.title('Odds Ratios with 95% CI')
plt.xlabel('Odds Ratio')
plt.ylabel('Metabolic Indicator')
plt.tight_layout()
plt.show()
```

Since the odds ratio for ALT is above 1, (indicated by the blue dot surpassing the red line), this shows that higher Alanin Aminotraferaz is greatly associated with the chance of gallstone disease.

Since LDL, HDL, ALP, Glucose, TC, and Triglyceride do not surpass 1, they are not great indicators of Gallstone disease.

```python

body_composition_features = [
'Body Mass Index (BMI)',
'Total Body Water (TBW)',
'Extracellular Water (ECW)',
'Intracellular Water (ICW)',
'Extracellular Fluid/Total Body Water (ECF/TBW)',
'Total Body Fat Ratio (TBFR) (%)',
'Lean Mass (LM) (%)',
'Body Protein Content (Protein) (%)',
'Visceral Fat Rating (VFR)',
'Bone Mass (BM)',
'Muscle Mass (MM)',
'Obesity (%)',
'Total Fat Content (TFC)',
'Visceral Fat Area (VFA)',
'Visceral Muscle Area (VMA) (Kg)',
'Hepatic Fat Accumulation (HFA)'
]


X_body = df[body_composition_features]
y = df['Gallstone Status']


import statsmodels.api as sm
X_body = sm.add_constant(X_body)
```

```python
logit_model_body = sm.Logit(y, X_body).fit()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


odds_ratios = np.exp(logit_model_body.params)
conf = np.exp(logit_model_body.conf_int())
p_values = logit_model_body.pvalues


results_df = pd.DataFrame({
'Variable': odds_ratios.index,
'Odds Ratio': odds_ratios.values,
'CI Lower': conf[0].values,
'CI Upper': conf[1].values,
'P-Value': p_values.values
})


results_df = results_df.sort_values(by='Odds Ratio', ascending=False)

# Plot
plt.figure(figsize=(9, 6))
sns.pointplot(
data=results_df,
y='Variable',
x='Odds Ratio',
color='green',
linestyles='none'
)


plt.errorbar(
x=results_df['Odds Ratio'],
y=np.arange(len(results_df)),
xerr=[
results_df['Odds Ratio'] - results_df['CI Lower'],
results_df['CI Upper'] - results_df['Odds Ratio']
],
fmt='none',
ecolor='gray',
capsize=3
)


plt.axvline(x=1, color='red', linestyle='--', linewidth=1)

plt.title('Odds Ratios with 95% CI for Body Composition Predictors of Gallstone Disease')
plt.xlabel('Odds Ratio')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

As you may notice, none of the green dots surpass the red line, meaning No body composition indicators are significant indicators of Gallstone Disease.
